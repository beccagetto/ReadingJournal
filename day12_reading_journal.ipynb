COMPARING CARDS:
- for built-in types: relational operators (<,>,==,stc.)
- for user-defined types: method def __cmp__(self, other): which returns a positive number if self is larger and a negative number if other is larger and 0 if they are equal
- for cards, need to pick if suit or rank is more important


```python
class Card(object):
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank

    def __cmp__(self, other):
        """ Compares 2 cards and returns 1 if first is greater, -1 if second is greater and 0 if they are tied"""
        # check the suits first
        if self.suit > other.suit: return 1
        if self.suit < other.suit: return -1

        # suits are the same... check ranks
        if self.rank > other.rank: return 1
        if self.rank < other.rank: return -1

        # ranks are the same... it's a tie
        return 0
```
